```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Basic statistics for numeric columns
print("\nBasic statistics for numeric columns:")
print(df.describe())

# Unique values and distribution for categorical columns
print("\nUnique values for each categorical column:")
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].nunique()} unique values")

# Visualize distribution of disaster types
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='incident_type', order=df['incident_type'].value_counts().index)
plt.title('Frequency of Different Incident Types')
plt.xlabel('Count')
plt.ylabel('Incident Type')
plt.show()


# Distribution of disasters by state
plt.figure(figsize=(18, 8))
sns.countplot(data=df, y='state', order=df['state'].value_counts().index)
plt.title('Number of Disasters by State')
plt.xlabel('Count')
plt.ylabel('State')
plt.show()

# Analyze month-wise distribution of disasters
if 'month' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='month', order=sorted(df['month'].unique()))
    plt.title('Monthly Distribution of Natural Disasters')
    plt.xlabel('Month')
    plt.ylabel('Number of Disasters')
    plt.show()

# Filter numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
numeric_df = df[numeric_columns]

# Generate correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap for Numeric Features')
plt.show()


# Trends over time (if there is a date or year column)
if 'year' in df.columns:
    yearly_disasters = df.groupby('year').size()
    plt.figure(figsize=(12, 6))
    yearly_disasters.plot(kind='line', marker='o')
    plt.title('Trend of Disasters Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Number of Disasters')
    plt.grid()
    plt.show()
```

```python
# Get all unique states and their count
unique_states = df["state"].unique()
num_states = len(unique_states)

print("Unique states:", unique_states)
print("Number of unique states:", num_states)
```

```python
df['incident_end_date'].fillna('Unknown', inplace=True)
```

```python
# Convert date columns to datetime, coercing errors to NaT
df['incident_begin_date'] = pd.to_datetime(df['incident_begin_date'], errors='coerce')
df['incident_end_date'] = pd.to_datetime(df['incident_end_date'], errors='coerce')
df['declaration_date'] = pd.to_datetime(df['declaration_date'], errors='coerce')

# Extract year and month after conversion
df['year'] = df['incident_begin_date'].dt.year
df['month'] = df['incident_begin_date'].dt.month

```

```python
df = df.dropna(subset=['incident_begin_date'])
print(df['incident_begin_date'].isna().sum())
```

```python
invalid_dates = df[df['incident_begin_date'].isna()]
print(invalid_dates[['incident_begin_date', 'incident_end_date', 'declaration_date']])
```

```python
state_disaster_count = df.groupby(['state', 'year', 'incident_type']).size().reset_index(name='disaster_count')
df = pd.get_dummies(df, columns=['incident_type'], drop_first=True)
```

```python
# List of mainland U.S. states
mainland_states = [
    "AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", 
    "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
    "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
    "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
    "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"
]

# Filter rows where the 'state' column is in mainland_states
df = df[df['state'].isin(mainland_states)]
```

```python
# List of columns to drop
columns_to_drop = [
    "fema_declaration_string", "fips", "place_code", "hash", "id", "last_refresh", 
    "last_ia_filing_date", "fy_declared", "ih_program_declared", "ia_program_declared", 
    "declaration_type", "pa_program_declared", "hm_program_declared", 
    "disaster_closeout_date", "declaration_request_number"  # Removed trailing space
]

# Ensure no trailing/leading spaces in column names
df.columns = df.columns.str.strip()

# Drop columns that exist in the DataFrame
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1)

# Display remaining columns for verification
print("Remaining columns:", df.columns)

```
